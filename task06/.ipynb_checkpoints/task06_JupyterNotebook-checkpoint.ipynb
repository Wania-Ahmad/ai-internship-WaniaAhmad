## Part A – Dataset Selection
### Dataset Description
- **Source:** Kaggle – Titanic: Machine Learning from Disaster  
- **Purpose:** Analyze demographic and socio‑economic factors influencing survival.  
- **Number of Records:** 891 rows  
- **Features:** PassengerId, Survived, Pclass, Name, Sex, Age, SibSp, Parch, Ticket, Fare, Cabin, Embarked  
- **Target Variable:** Survived (0 = did not survive, 1 = survived)

```python
# Load dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

# Preview first 5 rows
df.head()

## Part B – Exploratory Data Analysis

### Structure & Summary
#Display dataset info and summary statistics
df.info()
df.describe(include="all")

### Missing Values & Duplicates
#Check for missing values and duplicates
df.isnull().sum()
df.duplicated().sum()

### Visualizations
#1. Survival Count
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
#Interpretation: More passengers did not survive than survived.

#2. Survival by Gender
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
#Interpretation: Women had a much higher survival rate compared to men.

#3. Age Distribution
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
#Interpretation: Most passengers were between 20–40 years old.

#4. Survival by Passenger Class
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Class")
plt.show()
#Interpretation: First‑class passengers had higher survival rates.

#5. Fare vs Survival
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare vs Survival")
plt.show()
#Interpretation: Passengers who paid higher fares were more likely to survive.
